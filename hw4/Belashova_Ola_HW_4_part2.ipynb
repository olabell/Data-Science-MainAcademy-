Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
sex = input("Введіть стать спортсмена ('f' - якщо дівчина або 'm' - якщо хлопець): ")
age = float(input("Введіть вік спортсмена: "))

if sex == 'f' or sex == 'm':
    if age > 0 and age <= 100:
        if sex == 'f'and age >= 18:
            print('Спортсмен-дівчина може брати участь у змаганні')
        elif sex == 'f'and age < 18:
            print(f'Спортсмен-дівчина не може брати участь у змаганні, бо їй лише {age} років')
        if sex == 'm'and age >= 16:
            print('Cпортсмен-хлопець може брати участь у змаганні')
        elif sex == 'm'and age < 16:
            print(f'Cпортсмен-хлопець не може брати участь у змаганні, бо йому лише {age} років')
    else:
        print('Некоректний вік')    
else:
    print('Стать введена неправильно')
```

### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруємо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

print(list_with_random_numbers)
                    
```

1) рохраховує мінімальне та максимальне значення

```python
#варіант через цикл
min_value = list_with_random_numbers[0]
max_value = list_with_random_numbers[0]

for num in list_with_random_numbers:
    if num < min_value:
        min_value = num

for num in list_with_random_numbers:
    if num > max_value:
        max_value = num
        
print("Мінімальне значення:", min_value)
print("Максимальне значення:", max_value)

#або спрощений варіант
print('______________________\n')
print("Мінімальне значення:", min(list_with_random_numbers))
print("Максимальне значення:", max(list_with_random_numbers))
```

#### 2. Cереднє значення послідовності

AVERAGE = сума значень / кількість значень

```python
average_value = sum(list_with_random_numbers) / len(list_with_random_numbers)
print("Середнє значення послідовності:", average_value)


# другий варіант
print('______________________\n')
import statistics
print("Середнє значення послідовності: ", statistics.mean(list_with_random_numbers))
```

#### 3. Медіана послідовності

Медіана – середнє число із групи чисел

```python
list_with_random_numbers.sort()
mid = len(list_with_random_numbers) // 2
res = (list_with_random_numbers[mid] + list_with_random_numbers[~mid]) / 2
print("Медіана послідовності : ", str(res))


# другий варіант
print('______________________\n')
print("Медіана послідовності : ", statistics.median(list_with_random_numbers))
```

4. частоту кожної цифри в послідовності

```python
frequency = {}
for number in list_with_random_numbers:
    count = frequency.get(number,0)
    frequency[number] = count + 1
    
frequency_list = frequency.keys()
for number in frequency_list:
    print(f' Число {number} має частоту {frequency[number]}')
    
```

#### 5. Cереднє квадратичне відхилення
Для вибірки з n елементів:
- Обчислюємо середнє арифметичне (mean) для всіх елементів у вибірці.
- Обчислюємо різницю між кожним елементом і середнім арифметичним.
- Підносимо кожну різницю до квадрату.
- Знаходимо суму отриманих квадратів різниць.
- Ділимо суму квадратів різниць на n - 1 (де n - кількість елементів у вибірці).
- Беремо квадратний корінь з отриманого значення для знаходження середньоквадратичного відхилення.

```python
def sample_standard_deviation(data):

    n = len(data)   #Обчислюємо довжину вибірки
    mean = sum(data) / n #Обчислюємо середнє арифметичне

    squared_deviations = [] #пустий список для збереження квадратів відхилень
    for x in data:
        squared_deviations.append((x - mean) ** 2) #Для кожного елементу рахуємо квадрат відхилення від середнього значення mean
    
    sample_std_deviation = (sum(squared_deviations) / (n - 1)) ** 0.5 #Обчислюємо стандартне відхилення
    return sample_std_deviation

result = sample_standard_deviation(list_with_random_numbers)
print("Cереднє квадратичне відхилення:", result)




# другий варіант
print('______________________\n')
print("Cереднє квадратичне відхилення: ", statistics.stdev(list_with_random_numbers))
```
