### Task 1
Create a visualization similar to the figure 
 - visualize the first 3 graphs using values from df
 - the last one is a direct proportionality

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

### Solution 1

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10 ))

axes[0, 0].plot(df['a'])
axes[0, 0].set_xlabel('Index')
axes[0, 0].set_ylabel('Value')
axes[0, 0].set_title('Line Plot for Column a')

axes[0, 1].plot(df['b'])
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Value')
axes[0, 1].set_title('Line Plot for Column b')

axes[1, 0].plot(df['c'])
axes[1, 0].set_xlabel('Index')
axes[1, 0].set_ylabel('Value')
axes[1, 0].set_title('Line Plot for Column c')

x = np.linspace(0, 29, 30)
y = x
axes[1, 1].plot(x, y)
axes[1, 1].set_xlabel('X')
axes[1, 1].set_ylabel('Y')
axes[1, 1].set_title('Linear Relationship (y = x)')

plt.tight_layout()
plt.show()
```

## Task 6
Create a visualization similar to the image below 
- for column `a` use red color and line of format `-.`
- for column `b` use orange color and thickness `10`
- for column `c` use yellow color and thickness `1` and marker `o`


### Solution 6

```python
sns.set_style("darkgrid")

plt.figure(figsize=(10, 6))

plt.plot(df['a'], color='red', linestyle='-.')
plt.plot(df['b'], color='orange', linewidth=10)
plt.plot(df['c'], color='yellow', marker='o', markersize=4, linewidth=1)

plt.xlabel('Index')
plt.ylabel('Value')

plt.show()
```

### Task 7
Create a visualization similar to the image below 

### Solution 7

```python
sns.set_style("darkgrid")

fig, axes = plt.subplots(3, 1, figsize=(15, 10))

axes[0].plot(df['a'], label='Line A')
axes[0].legend(loc='lower right')

axes[1].plot(df['b'], label='Line B')
axes[1].legend(loc='center left')

axes[2].plot(df['c'])

plt.show()
```

### Task 8
Create a visualization similar to the image below 
- use a tomato color and a column spacing of 0.7

### Solution 8

```python
plt.figure(figsize=(11, 5))
plt.bar(df.index, df['a'], color='#f5a69c', edgecolor='none', width=0.7, label='a')
plt.legend()
plt.show()
```

### Task 9
Create a visualization similar to the image below 
- add all signatures and the correct marker

### Solution 9

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12, 5))

plt.plot(df['a'], linewidth=1, marker='^', markersize=3)

# Set labels and title
plt.xlabel('This is the X Axis', fontsize=13)
plt.ylabel('This is the Y Axis', fontsize=6)
plt.title('This is the Title', fontsize=8)



plt.xticks(range(0, 30, 1))
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)

plt.show()
```
